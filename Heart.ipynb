```python
# Chargement et étude rapide du jeu de données Pima
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Charger les données
Pima_DF = pd.read_csv('heart.dat', sep=' ', 
                      names=['age','sex','chest pain','blood pressure','cholestoral',
                             'blood sugar','electrocardiographic','heart rate achieved',
                             'induced angina','oldpeak','the slope of the peak','vessels',
                             'thal','target'])

Pima_DF.head()
```

```python
df = Pima_DF.copy()
```

```python
data = df.values
```

```python
X = df[df.columns[:-1]].values
X.shape
```

```python
y = df['target'].values
y.shape
```

```python
y = y.reshape(y.shape[0], 1)
y.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
stdScalerX = StandardScaler().fit(X)
X_std = stdScalerX.transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test, = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'n_neighbors': np.arange(1, 100),
             'metric' : ['euclidean', 'manhattan']}
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5)
```

```python
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
model = grid.best_estimator_
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, model.predict(X_test))
```

```python
from sklearn.model_selection import learning_curve
```

```python
N, train_score, val_score = learning_curve(model, X_train, y_train, train_sizes=np.linspace(0.1, 1, 10), cv=5)
```

```python
print(N)
plt.plot(N, train_score.mean(axis=1), label='train')
plt.plot(N, val_score.mean(axis=1), label='validation')
plt.xlabel('train_size')
plt.legend()
```
